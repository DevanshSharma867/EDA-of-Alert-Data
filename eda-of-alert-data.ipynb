### Configuration and Setup

This section defines key configuration variables such as file paths and date columns. Proper configuration ensures that the analysis is flexible, reproducible, and easy to adapt to new datasets or environments.

```python
DATA_PATH = '/Users/yourusername/Downloads/Employee_Attrition.csv'
DATE_COLUMNS = ['CREATED_ON', 'UPDATED_ON', 'CLOSED_ON', 'ACK_TIME']
DAY_ORDER = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

### Data Loading Functionality

A dedicated function is used to load the dataset, parse date columns, and handle potential errors. This approach ensures that the data is consistently and reliably prepared for analysis, regardless of its source or format.


```python
def load_data(data_path=DATA_PATH):
    try:
        df = pd.read_csv(
            data_path,
            parse_dates=DATE_COLUMNS,
            dayfirst=True,
            engine='python',
            encoding='utf-8'
        )
        for col in DATE_COLUMNS:
            if col in df.columns:
                df[col] = pd.to_datetime(df[col], errors='coerce', utc=True)
        return df
    except Exception as e:
        print(f"❌ Data loading error: {str(e)}")
        return None
```

### Data Cleaning Process

Data cleaning is a critical step that addresses missing values, inconsistent formatting, and potential data quality issues. The cleaning function standardizes string columns and ensures the dataset is ready for analysis and modeling.

```python
def clean_data(df):
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    string_columns = df.select_dtypes(include=['object']).columns
    for col in string_columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace(['nan', 'None', ''], np.nan)
    return df
```

### Feature Engineering Explained

Feature engineering derives new variables from existing data, such as extracting time-based features or categorizing statuses. These engineered features can reveal hidden patterns and improve the effectiveness of subsequent analyses and models.

```python
def engineer_features(df):
    if 'CREATED_ON' in df.columns:
        df['CREATED_HOUR'] = df['CREATED_ON'].dt.hour
        df['CREATED_DAY'] = df['CREATED_ON'].dt.day_name()
        df['CREATED_MONTH'] = df['CREATED_ON'].dt.month
        df['CREATED_DATE'] = df['CREATED_ON'].dt.date
    if 'CLOSED_ON' in df.columns and 'CREATED_ON' in df.columns:
        mask = (df['CLOSED_ON'].notna()) & (df['CREATED_ON'].notna())
        df.loc[mask, 'RESOLUTION_TIME_HOURS'] = (
            (df.loc[mask, 'CLOSED_ON'] - df.loc[mask, 'CREATED_ON']).dt.total_seconds() / 3600
        ).clip(lower=0, upper=8760)
    if 'ACK_TIME' in df.columns and 'CREATED_ON' in df.columns:
        mask = (df['ACK_TIME'].notna()) & (df['CREATED_ON'].notna())
        df.loc[mask, 'ACK_DELAY_MINUTES'] = (
            (df.loc[mask, 'ACK_TIME'] - df.loc[mask, 'CREATED_ON']).dt.total_seconds() / 60
        ).clip(lower=0, upper=10080)
    if 'STATUS' in df.columns:
        def _categorize_status(status):
            if pd.isna(status) or status == 'nan':
                return 'Unknown'
            status_lower = str(status).lower()
            if 'open' in status_lower or 'new' in status_lower:
                return 'Active'
            elif 'closed' in status_lower or 'resolved' in status_lower:
                return 'Resolved'
            elif 'progress' in status_lower or 'assigned' in status_lower:
                return 'In Progress'
            else:
                return 'Other'
        df['STATUS_CATEGORY'] = df['STATUS'].apply(_categorize_status)
    return df
```

### Pattern Analysis Overview

Pattern analysis explores key trends and distributions in the alert data, such as severity, status, temporal patterns, and resolution times. Visualizations and metrics from this section provide actionable insights for operational improvements.

```python
def analyze_patterns(df):
    metrics = {}
    try:
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Alert System Patterns Analysis', fontsize=16, fontweight='bold')
        if 'SEVERITY' in df.columns:
            severity_data = df['SEVERITY'].dropna()
            if len(severity_data) > 0:
                severity_counts = severity_data.value_counts()
                axes[0,0].pie(severity_counts.values, labels=severity_counts.index, autopct='%1.1f%%', startangle=90)
                axes[0,0].set_title('Alert Severity Distribution')
        if 'STATUS_CATEGORY' in df.columns:
            status_data = df['STATUS_CATEGORY'].dropna()
            if len(status_data) > 0:
                status_counts = status_data.value_counts()
                status_counts.plot(kind='bar', ax=axes[0,1])
                axes[0,1].set_title('Alert Status Categories')
                axes[0,1].tick_params(axis='x', rotation=45)
        if 'CREATED_HOUR' in df.columns:
            hourly_data = df['CREATED_HOUR'].dropna()
            if len(hourly_data) > 0:
                hourly_counts = hourly_data.value_counts().sort_index()
                axes[0,2].plot(hourly_counts.index, hourly_counts.values, marker='o')
                axes[0,2].set_title('Alerts by Hour of Day')
                axes[0,2].set_xlabel('Hour')
                axes[0,2].set_ylabel('Alert Count')
        if 'CREATED_DAY' in df.columns:
            day_data = df['CREATED_DAY'].dropna()
            if len(day_data) > 0:
                day_counts = day_data.value_counts().reindex(DAY_ORDER, fill_value=0)
                axes[1,0].bar(range(len(day_counts)), day_counts.values)
                axes[1,0].set_xticks(range(len(day_counts)))
                axes[1,0].set_xticklabels(day_counts.index, rotation=45)
                axes[1,0].set_title('Alerts by Day of Week')
        if 'RESOLUTION_TIME_HOURS' in df.columns:
            resolution_times = df['RESOLUTION_TIME_HOURS'].dropna()
            if len(resolution_times) > 0:
                q99 = resolution_times.quantile(0.99)
                filtered_times = resolution_times[resolution_times <= q99]
                axes[1,1].hist(filtered_times, bins=30, alpha=0.7, edgecolor='black')
                axes[1,1].set_title('Resolution Time Distribution (99th percentile)')
                axes[1,1].set_xlabel('Hours')
                axes[1,1].set_ylabel('Frequency')
        if 'CREATED_ON' in df.columns:
            valid_dates = df['CREATED_ON'].dropna()
            if len(valid_dates) > 0:
                monthly_data = df[df['CREATED_ON'].notna()].set_index('CREATED_ON')
                monthly_counts = monthly_data.resample('M').size()
                axes[1,2].plot(monthly_counts.index, monthly_counts.values, marker='o')
                axes[1,2].set_title('Monthly Alert Trend')
                axes[1,2].tick_params(axis='x', rotation=45)
        plt.tight_layout()
        plt.show()
        if 'CREATED_HOUR' in df.columns:
            hourly_data = df['CREATED_HOUR'].dropna()
            if len(hourly_data) > 0:
                metrics['peak_hour'] = int(hourly_data.value_counts().idxmax())
        if 'RESOLUTION_TIME_HOURS' in df.columns:
            resolution_data = df['RESOLUTION_TIME_HOURS'].dropna()
            if len(resolution_data) > 0:
                metrics['avg_resolution_hours'] = round(float(resolution_data.mean()), 2)
        if 'SEVERITY' in df.columns:
            severity_data = df['SEVERITY'].dropna()
            if len(severity_data) > 0:
                metrics['most_common_severity'] = str(severity_data.value_counts().idxmax())
    except Exception as e:
        print(f"⚠️ Error in pattern analysis: {str(e)}")
    return metrics
```

### Performance Analysis

Performance analysis evaluates key metrics such as resolution time, acknowledgment delay, and alert status distribution. These insights help organizations monitor service levels, identify bottlenecks, and optimize alert management processes.

```python
def analyze_performance(df):
    kpis = {}
    try:
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Alert System Performance Metrics', fontsize=16, fontweight='bold')
        if 'RESOLUTION_TIME_HOURS' in df.columns and 'SEVERITY' in df.columns:
            valid_data = df[(df['RESOLUTION_TIME_HOURS'].notna()) & (df['SEVERITY'].notna())]
            if len(valid_data) > 0:
                resolution_by_severity = valid_data.groupby('SEVERITY')['RESOLUTION_TIME_HOURS'].agg(['mean', 'median', 'std']).round(2)
                if len(resolution_by_severity) > 0:
                    resolution_by_severity['mean'].plot(kind='bar', ax=axes[0,0])
                    axes[0,0].set_title('Average Resolution Time by Severity')
                    axes[0,0].set_ylabel('Hours')
                    axes[0,0].tick_params(axis='x', rotation=45)
        if 'ACK_DELAY_MINUTES' in df.columns:
            ack_delays = df['ACK_DELAY_MINUTES'].dropna()
            if len(ack_delays) > 0:
                q95 = ack_delays.quantile(0.95)
                filtered_acks = ack_delays[ack_delays <= q95]
                axes[0,1].hist(filtered_acks, bins=30, alpha=0.7, edgecolor='black')
                axes[0,1].set_title('Acknowledgment Delay Distribution')
                axes[0,1].set_xlabel('Minutes')
                axes[0,1].set_ylabel('Frequency')
        if 'STATUS' in df.columns:
            status_data = df['STATUS'].dropna()
            if len(status_data) > 0:
                status_counts = status_data.value_counts().head(10)
                axes[1,0].barh(range(len(status_counts)), status_counts.values)
                axes[1,0].set_yticks(range(len(status_counts)))
                axes[1,0].set_yticklabels(status_counts.index)
                axes[1,0].set_title('Top 10 Alert Statuses')
        if 'CREATED_HOUR' in df.columns and 'CREATED_DAY' in df.columns:
            valid_data = df[(df['CREATED_HOUR'].notna()) & (df['CREATED_DAY'].notna())]
            if len(valid_data) > 0:
                heatmap_data = valid_data.groupby(['CREATED_DAY', 'CREATED_HOUR']).size().unstack(fill_value=0)
                if not heatmap_data.empty:
                    heatmap_data = heatmap_data.reindex(DAY_ORDER, fill_value=0)
                    heatmap_data = heatmap_data.fillna(0).replace([np.inf, -np.inf], 0)
                    if heatmap_data.sum().sum() > 0:
                        sns.heatmap(heatmap_data, ax=axes[1,1], cmap='YlOrRd', cbar_kws={'label': 'Alert Count'}, robust=True)
                        axes[1,1].set_title('Alert Volume Heatmap (Day vs Hour)')
                    else:
                        axes[1,1].text(0.5, 0.5, 'Insufficient data for heatmap', ha='center', va='center', transform=axes[1,1].transAxes)
        plt.tight_layout()
        plt.show()
        if 'RESOLUTION_TIME_HOURS' in df.columns:
            resolution_data = df['RESOLUTION_TIME_HOURS'].dropna()
            if len(resolution_data) > 0:
                kpis['mttr_hours'] = round(float(resolution_data.mean()), 2)
        if 'ACK_DELAY_MINUTES' in df.columns:
            ack_data = df['ACK_DELAY_MINUTES'].dropna()
            if len(ack_data) > 0:
                kpis['mtta_minutes'] = round(float(ack_data.mean()), 2)
        kpis['total_alerts'] = len(df)
        if 'CREATED_ON' in df.columns:
            valid_dates = df['CREATED_ON'].dropna()
            if len(valid_dates) > 0:
                date_range = (valid_dates.max() - valid_dates.min()).days
                if date_range > 0:
                    kpis['avg_daily_alerts'] = round(len(df) / date_range, 2)
        if 'STATUS_CATEGORY' in df.columns:
            status_data = df['STATUS_CATEGORY'].dropna()
            if len(status_data) > 0:
                resolved_count = (status_data == 'Resolved').sum()
                kpis['resolution_rate_percent'] = round((resolved_count / len(status_data)) * 100, 2)
    except Exception as e:
        print(f"⚠️ Error in performance analysis: {str(e)}")
    return kpis
```

### Summary Reporting

The summary reporting section consolidates key findings, metrics, and recommendations from the EDA. It provides an executive overview for stakeholders and supports data-driven decision-making.

```python
def print_summary(df, overview, patterns, performance):
    print("\n" + "="*60)
    print("📊 ALERT SYSTEM ANALYSIS - EXECUTIVE SUMMARY")
    print("="*60)
    print(f"📈 Total Alerts Analyzed: {performance.get('total_alerts', len(df)):,}")
    print(f"📅 Analysis Period: {overview.get('date_range', 'N/A')}")
    print(f"🎯 Data Quality Score: {overview.get('data_quality_score', 'N/A')}%")
    print("\n🔍 KEY FINDINGS:")
    if 'peak_hour' in patterns:
        print(f"   1. Peak alert activity occurs at {patterns['peak_hour']}:00")
    if 'avg_resolution_hours' in patterns:
        print(f"   2. Average resolution time is {patterns['avg_resolution_hours']} hours")
    if 'resolution_rate_percent' in performance:
        print(f"   3. Overall resolution rate is {performance['resolution_rate_percent']}%")
    print("\n💡 RECOMMENDATIONS:")
    recommendations = [
        "Implement automated alert correlation to reduce noise",
        "Establish SLA targets based on severity levels",
        "Consider staffing adjustments during peak hours",
        "Implement proactive monitoring for recurring issues",
        "Establish alert fatigue monitoring and mitigation"
    ]
    for i, rec in enumerate(recommendations, 1):
        print(f"   {i}. {rec}")
    print("="*60)
```

### Main Function and Pipeline Execution

The main function orchestrates the entire EDA pipeline, from data loading and cleaning to feature engineering, analysis, and reporting. This modular approach ensures reproducibility and makes it easy to adapt the workflow for new datasets or business requirements.

```python
def main():
    df = load_data(DATA_PATH)
    if df is None:
        print("Failed to load data. Check file path and format.")
        return
    df = clean_data(df)
    df = engineer_features(df)
    overview = {
        'total_records': len(df),
        'total_columns': len(df.columns),
        'memory_usage_mb': round(df.memory_usage(deep=True).sum() / 1024**2, 2),
        'date_range': f"{df['CREATED_ON'].min()} to {df['CREATED_ON'].max()}" if 'CREATED_ON' in df.columns else 'N/A',
        'data_quality_score': round(((df.size - df.isnull().sum().sum()) / df.size) * 100, 2)
    }
    patterns = analyze_patterns(df)
    performance = analyze_performance(df)
    print_summary(df, overview, patterns, performance)
    print("\n✅ EDA pipeline completed.")

    # --- Auto EDA using ydata_profiling ---
    try:
        from ydata_profiling import ProfileReport
        print("\n🤖 Generating automated EDA profile report...")
        sample_df = df.sample(n=min(10000, len(df)), random_state=42) if len(df) > 10000 else df
        profile = ProfileReport(sample_df, title="Auto EDA Report", explorative=True, minimal=True)
        profile.to_file("/kaggle/working/auto_eda_report.html")
        print("✅ Auto EDA HTML report saved as /kaggle/working/auto_eda_report.html")
    except Exception as e:
        print(f"⚠️ Auto EDA profiling failed: {str(e)}")

    # --- JSON Output ---
    output = {
        'overview': overview,
        'patterns': patterns,
        'performance': performance
    }
    with open("/kaggle/working/eda_output.json", "w", encoding="utf-8") as f:
        json.dump(output, f, indent=2, default=str)
    print("✅ JSON output saved as /kaggle/working/eda_output.json")

if __name__ == "__main__":
    main()
```
